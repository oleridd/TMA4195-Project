## Testing uniform random walk

```python
import sys
sys.path.append("../../")

from scipy.stats import lognorm
import matplotlib.pyplot as plt
import numpy as np

from Numeric.RandomWalkSingleBoundary import RandomWalkSingleBoundary
```

```python
step = lognorm(.01, -0.5)
```

```python
sample_size=10000

plt.hist(
    np.sort(step.rvs(size=sample_size))[:int(0.995*sample_size)],
    bins=100,
    density=True,
    histtype='stepfilled',
    );
```

```python
RW = RandomWalkSingleBoundary(N=1000, Nstep=100, step=step, z_boundary=10)
```

```python
RW.plot2D(timestep=1)
```

```python
RW.plot2D()
```
