## Testing uniform random walk

```python
import sys
sys.path.append("../../")

from scipy.stats import lognorm
import matplotlib.pyplot as plt
import numpy as np

from Numeric.RandomWalkSim import RandomWalk
```

```python
step = lognorm(0.1, 0)
```

```python
sample_size=10000

plt.hist(
    np.sort(step.rvs(size=sample_size))[:int(0.995*sample_size)],
    bins=100,
    density=True,
    histtype='stepfilled',
    );
```

```python
RW = RandomWalk(N=1000, Nstep=100, step=step, D=2)
```

```python
RW.plot2D(timestep=1)
```

```python
RW.plot2D()
```
